## Functions
make repeated tasks easier

````python
"""
This program calculated the area of a circle
"""

# Definition of a function
def findArea(radius): 
    PI = 3.142
    area = PI * (radius * radius)
    return area

# call function
print(findArea(5.5))
print(findArea(7))
````

````python
# list of radii
radius = [5,6,8,2,3,6]

for element in radius:
    area = findArea(element)
    print('The area of a circle with radius', element, 'is:', area)
````

````python
def DNA2RNA(DNA):
    """
    converts DNA string to RNA string.
    Parameters: DNA sequence as string
    Return: RNA sequence as a string
    """
    transliterate = DNA.maketrans("tT","uU")
    RNA = DNA.translate(transliterate)
    return RNA
````

Excercise 2: Implement a function that takes as input three variables and returns the largest of the three. 


````python
def max_of_three(var_1,var_2,var_3):
    highest = 0
    if  var_1>var_2 and var_1>var_3:
        highest = var_1
    elif var_2 > var_1 and var_2>var_3:
        highest = var_2
    else:
        highest = var_3
    return(f"The highest value is: {highest}")

max_of_three(15,45,1)
````

Write a program that asks the user how many Fibonnaci numbers to generate and then generates them


````python
def generate_fibo():
    count = int(input("How many numbers of the Fibonacci series do you want to generate?"))
    k = 1
    if count == 0: fib=[]
    elif count == 1: fib=[1]
    elif count == 2: fib=[1,1]
    else:
        fib=[1,1]
        while k<(count-1):
            fib.append(fib[k]+fib[k-1])
            k+=1
    return fib

generate_fibo()
````

Write a rock paper scissors game for two players. The choice of rock, paper, or scissor will be given via keyboard input.


````python
Player_1 = input("Player 1: Do you choose rock, paper, or scissors?")
Player_2 = input("Player 2: Do you choose rock, paper, or scissors?")

def rock_paper_scissors(u1,u2):
    if u1==u2:
        return("It is a tie, play again")
    elif u1=='rock':
        if u2=='scissors': return("Player 1 wins with: rock")
        else: return("Player 2 wins with: paper")
    elif u1=='paper':
        if u2=='rock': return("Player 1 wins with: paper")
        else: return("Player 2 wins with: scissors")
    elif u1=='scissors':
        if u2=='paper': return("Player 1 wins with: scissors")
        else: return("Player 2 wins with: rock")
    else: return("invalid input, I am missing at least one rock, paper, or scissor")

print(rock_paper_scissors(Player_1,Player_2))
````
