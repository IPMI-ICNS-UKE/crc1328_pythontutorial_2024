```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
def display(img,cmap=None):
    fig = plt.figure(figsize=(10,8))
    ax = fig.add_subplot(111)
    ax.imshow(img,cmap=cmap)
```

```python
cells = cv2.imread('data/cells3.png')
```

```python
display(cells)
```

```python
cells_blur = cv2.medianBlur(cells,5)
```

```python
display(cells_blur)
```

```python
ret, cells_thresh = cv2.threshold(cells_blur,50,255,cv2.THRESH_BINARY_INV)
```

```python
display(cells_thresh)
```

```python
contours, hierarchy = cv2.findContours(cells_thresh.copy(), cv2.RETR_CCOMP, cv2.CHAIN_APPROX_SIMPLE)
```
